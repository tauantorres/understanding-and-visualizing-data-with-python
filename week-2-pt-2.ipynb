### `Visualizing Data in Python`
#### `Tables, Histograms, Boxplots, and Slicing for Statistics`

When working with a new dataset, one of the most useful things to do is to begin to visualize the data. By using tables, histograms, box plots, and other visual tools, we can get a better idea of what the data may be trying to tell us, and we can gain insights into the data that we may have not discovered otherwise.

Today, we will be going over how to perform some basic visualisations in Python, and, most importantly, we will learn how to begin exploring data from a graphical perspective.

```python
# We first need to import the packages that we will be using
import seaborn as sns # For plotting
import matplotlib.pyplot as plt # For showing plots

# Load in the data set
tips_data = sns.load_dataset("tips")
```

#### `Visualizing the Data - Tables`
When you begin working with a new data set,  it is often best to print out the first few rows before you begin other analysis. This will show you what kind of data is in the dataset, what data types you are working with, and will serve as a reference for the other plots that we are about to make. 

```python
# Print out the first few rows of the data
tips_data.head(10)
```

#### `Describing Data`
Summary statistics, which include things like the mean, min, and max of the data, can be useful to get a feel for how large some of the variables are and what variables may be the most important. 

```python
# Print out the summary statistics for the quantitative variables
tips_data.describe()
```

#### `Creating a Histogram`

After we have a general 'feel' for the data, it is often good to get a feel for the shape of the distribution of the data.

```python
# Plot a histogram of the total bill
sns.histplot(data = tips_data, x = "total_bill", kde = False).set(title = "Histogram of Total Bill", xlabel = "Total Bill", ylabel = "Frequency");
plt.show()
```

```python
# Plot a histogram of the Tips only
sns.histplot(data = tips_data, x = "tip", kde = False).set(title = "Histogram of Tips", xlabel = "Tips", ylabel = "Frequency");
plt.show()
```

```python
# Plot a histogram of both the total bill and the tips'
sns.histplot(data = tips_data, x = "total_bill", kde = False, color = "red", alpha = 0.5)
sns.histplot(data = tips_data, x = "tip", kde = False, color = "blue", alpha = 0.5).set(title = "Histogram of Total Bill and Tips", xlabel = "Total Bill and Tips", ylabel = "Frequency");
plt.show()
```

#### `Creating a Boxplot`

Boxplots do not show the shape of the distribution, but they can give us a better idea about the center and spread of the distribution as well as any potential outliers that may exist. Boxplots and Histograms often complement each other and help an analyst get more information about the data

```python
# Create a boxplot of the total bill amounts
sns.boxplot(data = tips_data, x = "total_bill").set(title = "Boxplot of Total Bill", xlabel = "Total Bill");

plt.xticks(range(0, int(tips_data["total_bill"].max()) + 10, 3))
plt.show()
```

```python
# Create a boxplot of the tips amounts
sns.boxplot(data = tips_data, x = "tip", color = "red").set(title = "Boxplot of Tips", xlabel = "Tips")

plt.xticks(range(0, int(tips_data["tip"].max()) + 1, 1))
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crie uma figura com dois subplots
fig, axes = plt.subplots(2, 1, figsize=(8, 10))  # 2 linhas, 1 coluna de subplots

# Create a boxplot of the total bill amounts
sns.boxplot(data=tips_data, x="total_bill", ax=axes[0], color="blue")
axes[0].set(title="Boxplot of Total Bill", xlabel="Total Bill")
axes[0].set_xticks(range(0, int(tips_data["total_bill"].max()) + 10, 3))

# Create a boxplot of the tips amounts
sns.boxplot(data=tips_data, x="tip", ax=axes[1], color="red")
axes[1].set(title="Boxplot of Tips", xlabel="Tips")
axes[1].set_xticks(range(0, int(tips_data["tip"].max()) + 1, 1))

# Ajuste o espaço entre os subplots
plt.tight_layout()

# Exiba os gráficos
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crie uma figura com dois subplots
fig, axes = plt.subplots(2, 1, figsize=(8, 10))  # 2 linhas, 1 coluna de subplots

# Create a boxplot of the total bill amounts
sns.boxplot(data=tips_data, x="total_bill", ax=axes[0], color="blue")
axes[0].set(title="Boxplot of Total Bill", xlabel="Total Bill")
axes[0].set_xticks(range(0, 60, 3))



# Create a boxplot of the tips amounts
sns.boxplot(data=tips_data, x="tip", ax=axes[1], color="red")
axes[1].set(title="Boxplot of Tips", xlabel="Tips")
axes[1].set_xticks(range(0, 60, 3))

# Ajuste o espaço entre os subplots
plt.tight_layout()

# Exiba os gráficos
plt.show()
```

#### `Creating Histograms and Boxplots Plotted by Groups`

While looking at a single variable is interesting, it is often useful to see how a variable changes in response to another. Using graphs, we can see if there is a difference between the tipping amounts of smokers vs. non-smokers, if tipping varies according to the time of the day, or we can explore other trends in the data as well.

```python
# Create a boxplot and histogram of the tips grouped by smoking status
sns.boxplot(data = tips_data, x = "tip", y = "smoker").set(title = "Boxplot of Tips by Smoking Status", xlabel = "Tips", ylabel = "Smoker")

plt.xticks(range(0, int(tips_data["tip"].max()) + 1, 1))
plt.grid(True, axis="both", linestyle="-", alpha=0.5)
plt.show();
```

```python
# Create a boxplot and histogram of the tips grouped by time of day
sns.boxplot(data = tips_data, x = "tip", y = "time").set(title = "Boxplot of Tips by Time of Day", xlabel = "Tips", ylabel = "Time of Day")

plt.xticks(range(0, int(tips_data["tip"].max()) + 1, 1))
plt.grid(True, axis="both", linestyle="-", alpha=0.5)
plt.show();

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Defina o tamanho da figura global
plt.figure(figsize=(10, 15))

# Crie um FacetGrid para o histograma com o tamanho dos gráficos ajustado
g = sns.FacetGrid(tips_data, row="time", height=4, aspect=3)
g = g.map(plt.hist, "tip")

# Defina os rótulos dos eixos x e y em ambos os gráficos
g.set_axis_labels("Tips", "Frequency")

# Itere sobre os eixos para ativar a grade em ambos os gráficos
for ax in g.axes.flat:
    ax.grid(True, axis="both", linestyle="-", alpha=0.5)

# Defina os yticks para variar de 0 até 70, com um passo de 10
plt.yticks(range(0, 70, 10))

# Defina os xticks para variar de 0 até o valor máximo da gorjeta, com um passo de 1
plt.xticks(range(0, int(tips_data["tip"].max()) + 1, 1))

plt.show()
```

```python
# Create a boxplot and histogram of the tips grouped by the day
sns.boxplot(data=tips_data, x="tip", y="day").set(title="Boxplot of Tips by Day", xlabel="Tips", ylabel="Day")

plt.xticks(range(0, int(tips_data["tip"].max()) + 2, 1))
plt.grid(True, axis="both", linestyle="-", alpha=0.5)
plt.show();
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 15))


g = sns.FacetGrid(tips_data, row="day", height=4, aspect=3)
g = g.map(plt.hist, "tip")
g.set_axis_labels("Tips", "Frequency")

for ax in g.axes.flat:
    ax.grid(True, axis="both", linestyle="-", alpha=0.5)


plt.yticks(range(0, 70, 10))
plt.xticks(range(0, int(tips_data["tip"].max()) + 1, 1))
plt.show()
```
